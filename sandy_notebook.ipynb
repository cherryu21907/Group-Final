```python
import pandas as pd

tree = pd.read_csv("trees.csv")
tree.head()
```

^ shows data has been imported successfully


```python
pd.set_option("display.max_rows", None)
tree['neighborhood'].value_counts()
```

^ok, we have the data for total trees in each neighborhood
Next up is visualization and setting up a top 10 version of the list

```python
tree["neighborhood"].value_counts().sort_index().plot.pie(y='neighborhood', 
                                                  subplots=False, 
                                                  figsize=(8,4))
```

^bit difficult to read so here's an abbreviated version

```python
tree["neighborhood"].value_counts().head(10).sort_index().plot.pie(ylabel="", 
                                                  subplots=False, 
                                                  figsize=(8,4))

```

Conclusion:

Coming in at a whopping five thousand and seventy three trees, Squirrel Hill South is the clear front runner of best neighborhood vis a vis tree greenery.
